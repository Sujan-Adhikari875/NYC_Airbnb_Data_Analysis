```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/SQL DATA/AB_NYC_2019.csv/AB_NYC_2019.csv")
df
```

```python
 #summary of your DataFrame
df.info()
#statistical summary of DataFrame
df.describe() 
```

```python
#check null and duplicates value
df.isnull().sum(), df.duplicated().sum()
```

```python
#handle Missing values
df["host_name"] = df["host_name"].fillna("Unknown")
df["name"] = df["name"].fillna("Unknown")
df["last_review"] = pd.to_datetime(df["last_review"])
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
df.head()
```

```python
print(df.isnull().sum())
print(f"The total duplicates values is {df.duplicated().sum()}")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df.head()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr,annot= True, linecolor= "red")
plt.title("Correlation between All numeric VAlue", loc="center")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x= "number_of_reviews", y ="price")
plt.title("Number_of_reviews VS Price")
plt.show()
```

```python
df.to_csv("AB_NYC_2019_clean.csv", index=False)
```
